```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("covid_vaccine_statewise.csv")
```

```python
data.head()
```

```python
print(data.groupby("State")[["First Dose Administered"]].sum())
```

```python
print(data.groupby("State")[["Second Dose Administered"]].sum())
```

```python
first_dose_administered = data.groupby("State")[["First Dose Administered"]].sum()
second_dose_administered = data.groupby("State")[["Second Dose Administered"]].sum()
```

```python
plt.figure(figsize=(12,6))
plt.plot(first_dose_administered.index,first_dose_administered["First Dose Administered"],label="First Dose",marker="o")
plt.plot(second_dose_administered.index,second_dose_administered["Second Dose Administered"],label="Second Dose",marker="o")
plt.xlabel("State")
plt.ylabel("Doses Adminitered")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
